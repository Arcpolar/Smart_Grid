#### 3 Data Processing and Clean Up
##### 3.1 Loading Consumption Data
##### This project used Fu, C. & Kazmi, H.'s method of combining 112 files into one CSV file stored in a pivot format for best performance (2022)

```python
#import libraries
import os
import tqdm
import numpy as np
import pandas as pd
#set file path for first level data folder
datapath=r'Downloads/archive/'
#set file path for meter 1/2 hourly kWh data folder
meterdatapath = r'Downloads/archive/halfhourly_dataset/halfhourly_dataset/'
#create list of file names for meter 1/2 hourly kWh data files
filenamelist = os.listdir(meterdatapath)
#list first level folder file names
os.listdir(datapath)
```

Beside data in halfhourly_dataset folder, following file will be used: uk_bank_holidays.csv, wather_daily_darksky.csv, and informations_households.csv

```python
#set list to hold halfhourlydata
halfhourlydataset = []
#loop to read every file to one list
for filename in tqdm.tqdm(filenamelist):
    dftemp = pd.read_csv(os.path.join(meterdatapath, filename),index_col='tstp',parse_dates=True, low_memory=False)
    dftemp['filename'] = filename.split('.')[0]
    dftemp = dftemp.replace('Null', np.nan).dropna()
    dftemp['energy(kWh/hh)'] = dftemp['energy(kWh/hh)'].astype('float')
    halfhourlydataset.append(dftemp)
#load list to data frame  
halfhourlydataset = pd.concat(halfhourlydataset, axis=0)
halfhourlydataset
```

```python
#store the data in a new CSV file for next processing
dfoutput = halfhourlydataset.groupby(['tstp', 'LCLid'])['energy(kWh/hh)'].mean().unstack()
dfoutput.to_csv('housekWh.csv')
```

#### 4. Unsupervised Learning (Clustering)
#### For electricity utility to manage the grid, the customer’s load profile needs to be classified to see the electricity usage behaviour. An unsupervised learning Clustering model is an excellent idea for visiting the different customer electricity consumption behaviour. This project used Viola L. G.'s method of clustering electricity usage profiles with the pivot method and K-means method (2018).
##### 4.1 Single Meter ID MAC000009
##### Consumption data need to be summed up from half an hour data into hour data for the internal increase to hourly. Following Viola L. G.'s method of clustering electricity usage profile with the pivot method (2018)

```python
#Pull meter data from meter ID MAC000009
uldata = meterconsumptiondata[['MAC000009']]
uldata.columns = ['kWh']
uldata
```

```python
#sum half an hour data to hourly data
dfhourly = uldata.resample('H').sum()
dfhourly['hour'] = dfhourly.index.hour
dfhourly.index = dfhourly.index.date
#pivot data
dfpivot = dfhourly.pivot(columns='hour')
#drop NA
dfpivot = dfpivot.dropna()
print(dfpivot)
dfpivot.T.plot(figsize=(13,8), legend=False, color='blue', alpha=0.02)
```

##### 4.1.1 K-Mean Clustering

```python
#import KMeans
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

dfpivotK = dfpivot
X = dfpivotK.values.copy()
#scaling
sc = MinMaxScaler()
X = sc.fit_transform(X)
kmeans = KMeans(n_clusters=5)
cluster_found = kmeans.fit_predict(X)
cluster_found_sr = pd.Series(cluster_found, name='cluster')
dfpivotK = dfpivotK.set_index(cluster_found_sr, append=True )
fig, ax= plt.subplots(1,1, figsize=(18,10))
color_list = ['blue','red','green','black','yellow']
cluster_values = sorted(dfpivotK.index.get_level_values('cluster').unique())

for cluster, color in zip(cluster_values, color_list):
    dfpivotK.xs(cluster, level=1).T.plot(
        ax=ax, legend=False, alpha=0.01, color=color, label= f'Cluster {cluster}'
        )
    dfpivotK.xs(cluster, level=1).median().plot(
        ax=ax, color=color, alpha=0.9, ls='--'
    )

ax.set_xticks(np.arange(1,25))
ax.set_ylabel('kilowatts')
ax.set_xlabel('hour')
```

##### 4.1.2 BIRCH Clustering

```python
from sklearn.cluster import Birch

dfpivotB = dfpivot
X = dfpivotB.values.copy()
    
#scaling
sc = MinMaxScaler()
X = sc.fit_transform(X)
#create BIRCH Model
birchmodel = Birch(n_clusters=5,threshold=0.1,branching_factor=100)
#training
cluster_found = birchmodel.fit_predict(X)
cluster_found_sr = pd.Series(cluster_found, name='cluster')
dfpivotB = dfpivotB.set_index(cluster_found_sr, append=True )
fig, ax= plt.subplots(1,1, figsize=(18,10))
color_list = ['blue','red','green','black','yellow']
cluster_values = sorted(dfpivotB.index.get_level_values('cluster').unique())
#plot cluster
for cluster, color in zip(cluster_values, color_list):
    dfpivotB.xs(cluster, level=1).T.plot(
        ax=ax, legend=False, alpha=0.01, color=color, label= f'Cluster {cluster}'
        )
    dfpivotB.xs(cluster, level=1).median().plot(
        ax=ax, color=color, alpha=0.9, ls='--'
    )

ax.set_xticks(np.arange(1,25))
ax.set_ylabel('kilowatts')
ax.set_xlabel('hour')
```

##### 4.2 Average All Meter Consumptions and Apply K-Mean Cluster
###### Taking average all meter consumptions to each time hourly interval

```python
#Take average of all meters to hourly interval
houseAVG = meterconsumptiondata.mean(axis=1).to_frame()
houseAVG.columns = ['kWh']
```

```python
houseAVG
```

```python
#sum half an hour data to hourly data
dfhourly = houseAVG.resample('H').sum()
dfhourly['hour'] = dfhourly.index.hour
dfhourly.index = dfhourly.index.date
#pivot data
dfpivot = dfhourly.pivot(columns='hour')
#drop NA
dfpivot = dfpivot.dropna()
print(dfpivot)
dfpivot.T.plot(figsize=(13,8), legend=False, color='blue', alpha=0.02)
```

```python
dfpivotK = dfpivot
X = dfpivotK.values.copy()
#scaling
sc = MinMaxScaler()
X = sc.fit_transform(X)
#create kmeans model
kmeans = KMeans(n_clusters=3)
#train the model
cluster_found = kmeans.fit_predict(X)
cluster_found_sr = pd.Series(cluster_found, name='cluster')
dfpivotK = dfpivotK.set_index(cluster_found_sr, append=True )
fig, ax= plt.subplots(1,1, figsize=(18,10))
color_list = ['blue','red','green']
cluster_values = sorted(dfpivotK.index.get_level_values('cluster').unique())

for cluster, color in zip(cluster_values, color_list):
    dfpivotK.xs(cluster, level=1).T.plot(
        ax=ax, legend=False, alpha=0.01, color=color, label= f'Cluster {cluster}'
        )
    dfpivotK.xs(cluster, level=1).median().plot(
        ax=ax, color=color, alpha=0.9, ls='--'
    )

ax.set_xticks(np.arange(1,25))
ax.set_ylabel('kilowatts')
ax.set_xlabel('hour')
ax.legend()
```

##### 2.3 Average Each Meters Consumption for K-Means Cluster

```python
#ignore warnings
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
#create new data frame
housesc1 = pd.DataFrame()
#loop go through all the meter pivot
for meternumber in list(meterconsumptiondata):
    elec_houseAVG = meterconsumptiondata[[meternumber]]
    elec_houseAVG.columns = ['kWh']
    df_uci_hourly = elec_houseAVG.resample('H').sum()
    df_uci_hourly['hour'] = df_uci_hourly.index.hour
    df_uci_hourly.index = df_uci_hourly.index.date
    df_uci_pivot = df_uci_hourly.pivot(columns='hour')
    df_uci_pivot = df_uci_pivot.dropna()
    elec_house3 = df_uci_pivot.mean(axis=0).to_frame()
    elec_house3.columns = [meternumber]
    add_house = elec_house3.T
    housesc1 = housesc1.append(add_house)
housesc1
```

```python
housesc1.T.plot(figsize=(13,8), legend=False, color='blue', alpha=0.02)
```

```python
X = housesc1.values.copy()
#scaling
sc = MinMaxScaler()
X = sc.fit_transform(X)
#create kmeans model
kmeans = KMeans(n_clusters=5)
#train the model
cluster_found = kmeans.fit_predict(X)
cluster_found_sr = pd.Series(cluster_found, name='cluster')
#assigne cluster number back to data frame
housesc1 = housesc1.set_index(cluster_found_sr, append=True )
fig, ax= plt.subplots(1,1, figsize=(18,10))
color_list = ['blue','red','green','black','yellow']
cluster_values = sorted(housesc1.index.get_level_values('cluster').unique())
#assign plot colour and plot cluster
for cluster, color in zip(cluster_values, color_list):
    housesc1.xs(cluster, level=1).T.plot(
        ax=ax, legend=False, alpha=0.001, color=color, label= f'Cluster {cluster}'
        )
    housesc1.xs(cluster, level=1).median().plot(
        ax=ax, color=color, alpha=0.99, ls='--'
    )

ax.set_xticks(np.arange(1,25))
ax.set_ylabel('kilowatts')
ax.set_xlabel('hour')
```

##### 5.2 Combine All the Data Together

```python
#Use previouse processed average data
meterdataset = houseAVG.copy()
#Add new time columns date, weekday, hour
meterdataset['timestamp'] = meterdataset.index
meterdataset['date'] = pd.to_datetime(meterdataset.index.date)
meterdataset['weekday'] = meterdataset.index.weekday
meterdataset['hour'] = meterdataset.index.hour + meterdataset.index.minute/60
#add temperature information
meterdataset = meterdataset.merge(temperaturedata, left_index=True, right_index=True)
#add holiday information
meterdataset = meterdataset.merge(dfholiday, on='date', how='left')
#drop date column
meterdataset = meterdataset.drop('date',axis=1)
#add Holiday column name
meterdataset['Holiday'] = meterdataset['Holiday'].fillna(-1)
#set index
meterdataset.set_index('timestamp',inplace=True)
#delete NA
meterdataset = meterdataset.dropna()
meterdataset
```

##### 5.4 Predication Models Training and Testing (Ridge Model vs. LightGBM)
##### With the knowledge learned from Fu, C. & Kazmi, H.'s method of Linear Regression Model and LightBGM (2022), this project will use Linear Regression Ridge Model and LightBGM for model testing. And all 5,567 household meters data have been used for calculating the mean of the data in section 5.2.

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.linear_model import Ridge
import lightgbm as lgb
```

```python
#setup a printing function for print one month result for easy understanding
def testdata_plot(df):
    plt.figure(figsize=(15,2))
    plt.plot(df.loc['2013-12'].index, df.loc['2013-12','kWh'], '-g', alpha=0.6, label='kWh')
    plt.plot(df.loc['2013-12'].index, df.loc['2013-12','kWh_pred'], '--r', alpha=0.8, label='kWh_pred')
    plt.legend(loc='lower center', ncol=3, bbox_to_anchor=(0.5, -0.5))
    plt.show() 
    print("*******************************************************************************************************************")
    status = True
    return None
```

##### 5.4.1 Models Training Only With Weekday and Hours of the Day

```python
#make feature variables list
featurelist = ['weekday','hour']
#set training data from Jan 2013 to Oct 2013
traindata = meterdataset.loc['2013-01':'2013-10'].copy()
#set test data from Nov 2013 to Feb 2014
testdata = meterdataset.loc['2013-11':].copy()
#Ridge Model Training and testing
Ridge_model = Ridge()
Ridge_model.fit(traindata[featurelist], traindata['kWh'])
testdata['kWh_pred'] = Ridge_model.predict(testdata[featurelist])
errors = abs(testdata['kWh_pred'] - testdata['kWh'])
RSQUARED = r2_score(testdata['kWh'], testdata['kWh_pred'])
MAE = mean_absolute_error(testdata['kWh'], testdata['kWh_pred'])
#plot month of December
print("Model: Linear regression Ridge")
print("R\u00B2: "+str(round(RSQUARED,4)))
print("Mean Absolute Error: "+str(round(MAE,4)))
testdata_plot(testdata)
#LGBM traing and testing
LGB_model = lgb.LGBMRegressor()
LGB_model.fit(traindata[featurelist], traindata['kWh'])
testdata['kWh_pred'] = LGB_model.predict(testdata[featurelist])
errors = abs(testdata['kWh_pred'] - testdata['kWh'])
RSQUARED = r2_score(testdata['kWh'], testdata['kWh_pred'])
MAE = mean_absolute_error(testdata['kWh'], testdata['kWh_pred'])
#plot month of December
print("Model: LGBM")
print("R\u00B2: "+str(round(RSQUARED,4)))
print("Mean Absolute Error: "+str(round(MAE,4)))
testdata_plot(testdata)
```
